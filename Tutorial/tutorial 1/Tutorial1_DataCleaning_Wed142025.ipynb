## Step 6: Clean the data

Next, let's clean the data.
- State whether or not there are impossible ages.  Remove impossible ages and replace with NaN.
- State whether or not thare are implausible blood pressure values.  Replace implausible blood pressure values with NaN

It's easy to work with null values, so an easy way to filter the data for cleanness is to check for impossible or implausible data with a NaN or NA

You can ask if a datafrane has any missing values with:  
df.isna().sum()

You can drop missing values with:

We can clean the data based on impossible or implausible values.

*** Clean the variables based on filtering or querying the variable and replace the bad values with NaN  

You can use any python-based methods that you would like to, but one way is to try out thinking through:  
.loc[row_condition, column_name]`as a filter

- The first argument selects **rows**
- The second argument selects **columns**
- Assignment to NaN updates only those cells located

Use | for OR and & for AND when combining conditions on Pandas columns.

An example would be:  
df.loc[(df["Variable"] < Bound1) | (df["Variable"] > Bound2), "Age"] = np.nan

```python
df.loc[df['Age'] < 0, 'Age'] = np.nan
df.loc[(df['SystolicBP'] < 50) | (df['SystolicBP'] > 300), 'SystolicBP'] = np.nan
print(df.isna().sum())
dfcleaned = df.dropna()
print(dfcleaned)
```
