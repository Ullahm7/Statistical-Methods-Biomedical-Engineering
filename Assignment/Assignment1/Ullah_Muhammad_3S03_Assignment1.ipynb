```python
# ensure correct datatype (removal of summery rows in WDU) 
wtdUnits = wtd[wtd['WMU'] != 'Total']

# ensure NULL values removed 
x = wtdUnits['Total Harvest'].dropna()

plt.subplot(1, 2, 1)
sns.histplot(x, kde=True)
plt.title('Distribution of Total Deer Harvest')
plt.xlabel('Total Harvest')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.boxplot(x=x)
plt.title('Distribution of Total Deer Harvest')
plt.xlabel('Total Harvest')
plt.ylabel('Frequency')

plt.show()
```
