Function generated with the assistance of OpenAI's ChatGPT (GPT-5, OpenAI, September 2025)

Prompt: "Write a Python function to calculate the factorial of a number recursively."  

```python
def factorial(n):  
    if n == 0:    
        return 1  
    else:  
        return n * factorial(n-1)
```
