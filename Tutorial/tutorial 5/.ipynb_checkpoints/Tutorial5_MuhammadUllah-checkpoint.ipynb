```python
# Jarque–Bera Normality Tests

groups = {    "Dark Chocolate (DC)": DC,
    "Dark + Milk (DM)": DM,
    "Milk Chocolate (MC)": MC
}

print("Jarque–Bera Normality Tests\n")

for name, data in groups.items():
    jb_stat, jb_p = stats.jarque_bera(data)
    
    print(f"{name}")
    print(f"  JB statistic = {jb_stat:.4f}")
    print(f"  p-value      = {jb_p:.4f}")
    
    if jb_p <= 0.05:
        print("  → Evidence of non-normality (reject H0)\n")
    else:
        print("  → No evidence against normality (fail to reject H0)\n")
```

```python
# Levene’s Test for Equal Variances
levene_stat, levene_p = stats.levene(DC, DM, MC)

print(f"Levene statistic = {levene_stat:.4f}")
print(f"p-value          = {levene_p:.4f}")

if levene_p <= 0.05:
    print("→ Variances are significantly different (reject H0)")
else:
    print("→ No evidence of unequal variances (fail to reject H0)")
```

### Question:  What is the F-statistic calculation?

```python
# Find the F statistic
```

The F statistic is the group mean square divided by the error mean square. 

This statistic follows an F distribution with (a-1) and a(n-1) degrees of freedom. For our example, the critical F value (upper tail) for α = 0.05, (a-1) = 2, and a(n-1) = 33 is ~3.3.

# Comparisons to conclude statistically difference between pairs

What is the conclusion based on the post-ANOVA testing of the differences between pair-wise groups?

```python
alpha = 0.05

means = group_means.values
labels = list(group_means.index)
x = np.arange(len(labels))

# ---- 95% CI ----
tcrit = stats.t.ppf(1 - alpha/2, df_error)
SE_mean = np.sqrt(MSE / n)

ci_low = means - tcrit * SE_mean
ci_high = means + tcrit * SE_mean
yerr = np.vstack([means - ci_low, ci_high - means])

# ---- Fisher LSD ----
LSD = tcrit * np.sqrt(2 * MSE / n)

# ---- Determine significance matrix ----
k = len(means)
sig_matrix = np.zeros((k, k), dtype=bool)

for i, j in combinations(range(k), 2):
    if abs(means[i] - means[j]) > LSD:
        sig_matrix[i, j] = True
        sig_matrix[j, i] = True

# ---- Assign compact letters ----
letters = [''] * k
current_letter = 'a'

for i in range(k):
    if letters[i] == '':
        letters[i] += current_letter
        for j in range(i+1, k):
            if not sig_matrix[i, j]:
                letters[j] += current_letter
        current_letter = chr(ord(current_letter) + 1)

# ---- Plot ----
plt.figure()
plt.errorbar(x, means, yerr=yerr, fmt='o', capsize=6)
plt.xticks(x, labels)
plt.title("Treatment Means with 95% CI\nFisher LSD Grouping (α = 0.05)")
plt.xlabel("Chocolate Type Group")
plt.ylabel("Antioxidant Capacity (micromolar TroloxE)")

# Add letters above bars
for i in range(k):
    plt.text(x[i], ci_high[i] + 0.05*(max(ci_high)-min(ci_low)),
             letters[i], ha='center', fontsize=14)

plt.tight_layout()
plt.show()
```
