Next we will random take a sample of size 10 from the population.  We will make a small dataframe from the Series made of the sampled data.

```python
# Sample 10 rows randomly to make a custom DataFrame

#  You need to enter that you will radnomly choose a sample of size 10 from the Series that were created
sample_indices = np.random.choice(range(100), size=10, replace=False)

# Create custom DataFrame from sampled rows
df_sample = pd.DataFrame({
    "Names": names.iloc[sample_indices].values,
    "Ages": ages.iloc[sample_indices].values,
    "Hemoglobin (g/L)": hemoglobin.iloc[sample_indices].values
})

df_sample
```

# Data Visualization:  Barplot in Python with Seaborn

Step 1.  Adapt the following code cell to use Seaborn to make a barplot that shows the Mean Hemoglobin level by Age Group.  
Step 2.  Make sure that you have the estimator as the mean and errorbar as sd; make sure that the plot has a title and that the axes are labeled.  
Step 3.  Add a markdown cell and briefly write if you think there is a difference between the hemoglobin levels of the different age groups or not and why.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# https://seaborn.pydata.org/generated/seaborn.barplot.html

# Step 1: Add Age Group:  This creates an Under_40 and 40+ variable named Age Group that is added to the dataframe
df_sample["Age Group"] = np.where(df_sample["Ages"] < 40, "Under_40", "40+")

# Step 2: Could calculate mean Hemoglobin for each group if using maplotlib
mean_hg = df_sample.groupby("Age Group")["Hemoglobin (g/L)"].mean()

# Step 3: Bar Plot
# plt.bar(x=mean_hg.index, height=mean_hg)
________________(x=df_sample["Age Group"], y=df_sample["Hemoglobin (g/L)"],estimator="mean",errorbar="sd")

#Step 4: Show the data with stripplot
sns.stripplot(    data=df_sample,    x="Age Group",    y="Hemoglobin (g/L)",    jitter=True,    size=6,    color="black",    alpha=0.7)

# Step 5: Add titles and axes
plt.title("Mean Hemoglobin by Age Group")
plt.ylabel("Hemoglobin (g/L)")
plt.xlabel("Age Group")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
