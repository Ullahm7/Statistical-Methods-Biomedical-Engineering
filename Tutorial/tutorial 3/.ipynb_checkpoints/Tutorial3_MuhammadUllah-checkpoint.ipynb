# How to Interpret Confidence Intervals
Consider this: you want to know the average height of all McMaster students, but you are also very busy, so you do not have the time to measure every student. You collect a random sample of heights and calculate the mean. This mean is **most likely different from the true population mean**, so you create a 95% confidence interval to capture a range where the true mean may lie (we will get to *how* in a bit).

Imagine you keep taking random samples, each resulting in a slightly different mean. You calculate the 95% CI for each group of samples. Magically, McMaster releases the heights for every student, and you compare the true mean to your calculated CIs. You notice that the **mean is contained in only 95% of them**.
