# Series

# DataFrame

```python
data = {
    'Name': ['Alice', 'Bob'],
    'Age': [25, 30]
}

my_df = pd.DataFrame(data)
print(my_df['Name'][1])  # Output: Bob
print("\n") # adds a line
print(my_df['Name'])
```

```python
names = pd.Series(['Alice', 'Bob', 'Charlie'],name="Names")

ages = pd.Series([25, 30, 35],name="Ages")

# Combine into a DataFrame
df = pd.concat([names,ages],axis='columns')
df
```

# Structures are:

✅ Ordered (you can access by position)

✅ Mutable (you can change values)

✅ Iterable (you can loop through it)

```python
# Create a NumPy array
arr = np.array([10, 20, 30])

# ✅ Ordered: Access by index
print("First element (ordered):", arr[0])  # Output: 10

# ✅ Mutable: Modify an element
arr[1] = 99
print("After mutation:", arr)  # Output: [10 99 30]

# ✅ Iterable: Loop through elements
print("Iterating over array:")
for item in arr:
    print(item)

print("For if loops:")
for item in arr:
    if item % 2 == 0:
        print(item)
```

#  Index and Columns

```python
# Original data
df

# Add new column: Hemoglobin levels
Hg = pd.Series([135, 140, 150], name="Hemoglobin (g/L)")
df1=pd.concat([df,Hg],axis='columns')

# Show the updated DataFrame
print("DataFrame:")
print(df1)

# Show index
print("\nIndex:") # \n adds a line for output
print(df1.index)

# Show column labels
print("\nColumns:")
print(df1.columns)
```
